# Seaborn – Class 2: Categorical Plots


## What Are Categorical Plots?

Categorical plots visualize data where at least one variable is **categorical**  
(e.g., day of week, gender, class).

Seaborn gives several easy options to:
- Compare **counts** or **averages** across categories
- Understand spread & distribution


## barplot() vs countplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.barplot(x="day", y="size", data=tips)
plt.show()
```

```python
sns.barplot(x="sex", y="total_bill", data=tips)
plt.show()
```

```python
sns.countplot(x="sex", data=tips)
plt.show()
```

```python
sns.countplot(x="day", data=tips)
plt.show()

# Mostly used in machine learning for class Imbalance.
```

```python
sns.countplot(x='size', data=tips)
plt.show()
```

## boxplot()

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.boxplot(x="sex", y="total_bill", data=tips)
plt.show()
```

## stripplot() & swarmplot()

```python
sns.stripplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.show()
```

| Plot        | Best for |
|-------------|-----------|
| **barplot** | Compare average values across categories |
| **countplot** | Show how many items fall in each category |
| **boxplot** | See median, quartiles, and outliers |
| **violinplot** | Same as boxplot + shows distribution |
| **stripplot** | View every observation (good for small data) |
| **swarmplot** | Like stripplot but avoids overlap |
