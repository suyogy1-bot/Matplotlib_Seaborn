```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movies_dataset_500.csv")
```

## Q11. Remove duplicate rows

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

## Q12. Remove extra spaces from Title

```python
df['Title'] = df['Title'].str.strip()
```

## Q13. Convert Title to Proper Case

```python
df['Title'] = df['Title'].str.title()
```

## Q14. Correct Genre spellings

```python
df['Genre'].unique()
```

```python
genre_mapping = {
    "Scifi": "Sci-Fi",
    "Romnce": "Romance",
    "Acton": "Action"
}
```

```python
df['Genre'] = df['Genre'].replace(genre_mapping)
```

```python
df['Genre'].unique()
```

## Q15. Convert Country to upper case

```python
df['Country'] = df['Country'].str.upper()
```

## Q16. Convert Language to lower case

```python
df['Language'] = df['Language'].str.lower()
```

## Q17. Convert Rating to float

```python
df['Rating'].dtype
```

## Q18. Clean Duration column

```python
df['Duration']
```

```python
df['Duration'] = df['Duration'].str.replace('min','')
```

```python
df['Duration']
```

```python
df['Duration'] = pd.to_numeric(df['Duration'], errors = 'coerce')
```

```python
df['Duration']
```

## Q19. Clean Budget column

```python
df['Budget']
```

```python
df['Budget'] = df['Budget'].str.replace("$", "").str.replace("M", "")
```

```python
df['Budget']
```

```python
df['Budget'] = pd.to_numeric(df['Budget'], errors = 'coerce')
```

```python
df['Budget']
```

## Q20. Clean Revenue column

```python
df['Revenue']
```

```python
df['Revenue'] = df['Revenue'].str.replace(",", "")
```

```python
df['Revenue']
```

```python
df['Revenue'] = pd.to_numeric(df['Revenue'], errors = 'coerce')
```

```python
df['Revenue']
```

## Q21. Convert Release_Year to integer

```python
df['Release_Year'] # already an integer
```

## Q22. Convert Added_Date to datetime

```python
df['Added_Date']
```

```python
df['Added_Date'] = pd.to_datetime(df['Added_Date'], format = 'mixed')
```

```python
df['Added_Date']
```

## Q23. Delete rows where Title is missing

```python
df['Title'].isna().sum()
```

```python
df = df.dropna(subset=['Title'])
```

```python
df['Title'].isna().sum()
```

```python
df.shape
```

## Q24. Fill missing Rating with mean

```python
df['Rating'].isna().sum()
```

```python
rating_mean = df['Rating'].mean()
df['Rating'] = df['Rating'].fillna(rating_mean)
```

```python
df['Rating'].isna().sum()
```

## Q25. Fill missing Country with mode

```python
df['Country'].isna().sum()
```

```python
country_mode = df['Country'].mode()[0]
```

```python
df['Country'] = df['Country'].fillna(country_mode)
```

```python
df['Country'].isna().sum()
```

## Q26. Fill missing Budget with median

```python
df['Budget'].isna().sum()
```

```python
budget_median = df['Budget'].median()
df['Budget'] = df['Budget'].fillna(budget_median)
```

```python
df['Budget'].isna().sum()
```

## Q27. Check remaining null values

```python
df.isna().sum()
```

## Q28. Save cleaned dataset

```python
df.to_csv("cleaned_pandas_matplotlit.csv")
```

## Q30. Average rating by platform

## Q31. Bar chart of platform-wise movie count

```python
platform_counts = df['Platform'].value_counts()

plt.figure(figsize = (8, 3))
plt.bar(platform_counts.index , platform_counts.values, color = 'skyblue')
plt.xlabel('Platform')
plt.ylabel('Movie Count')
plt.title('Platform-wise movie count')
plt.show()
```

## Q32. Pie chart of genre distribution

```python
genre_counts = df['Genre'].value_counts()

plt.figure(figsize = (4,4))
plt.pie(
    genre_counts.values,
    labels = genre_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Genre Distribution')
plt.axis('equal')
plt.show()
```

## Q33. Histogram of Ratings

```python
plt.figure(figsize=(6,2))
plt.hist(df['Rating'], bins = 20, color = 'red', edgecolor = 'black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Histogram of Movie Ratings')
plt.show()
```

## Q34. Line chart of release year counts

```python
year_counts = df['Release_Year'].value_counts().sort_index()

plt.figure(figsize=(4,2))
plt.plot(year_counts.index, year_counts.values, marker='o', color='green')
plt.xlabel('Release Year')
plt.ylabel('Movie Count')
plt.title('Release Year Counts')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

## Q35. Box plot of Ratings

```python
plt.figure(figsize=(4,4))
plt.boxplot(df['Rating'])
plt.ylabel('Rating')
plt.title('Box Plot of Movie Ratings')
plt.show()
```

## Q36. Scatter plot of Budget vs Revenue

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Budget'], df['Revenue'], alpha=0.6, color='purple')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Scatter Plot of Budget vs Revenue')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Because you already cleaned both ‎`Budget` and ‎`Revenue` to numeric 
(with NaNs where missing), 
matplotlib will automatically ignore the NaNs 
and plot only valid points.
