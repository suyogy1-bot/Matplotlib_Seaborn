# Seaborn – Class 3: Distribution Plots

## What Are Distribution Plots?

Distribution plots show how data values are **spread or concentrated**.

Seaborn offers:
- **`histplot()`** → modern histogram (with optional KDE)
- **`kdeplot()`** → smooth curve of distribution
- **`rugplot()`** → small marks showing individual values  
*(Older `distplot()` is now replaced by `histplot()`.)*


## histplot() – Basic Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(tips['total_bill'], kde=True)
plt.show()
```

## kdeplot() – Kernel Density Estimate

```python
sns.kdeplot(tips['total_bill'], fill=True)
plt.show()
```

## distplot() (Legacy) vs histplot()

> `distplot()` is deprecated.
> Use `histplot()` or `kdeplot()` instead.


## rugplot() – Small Ticks for Observations

```python
sns.rugplot(tips['total_bill'])
plt.show()
```

## Multiple Distributions with hue

```python
sns.histplot(data=tips , x='total_bill', hue='time')
plt.show()

# Hue main hum woh column denge jisme categories aa rhi ho.
```

Key takeaways

histplot() → flexible histogram (can combine with KDE).

kdeplot() → smooth density curve.

rugplot() → quick look at raw values.

distplot() is deprecated → prefer histplot().

Use hue to compare groups in the same distribution plot.

```python
# 1. Histogram on left and KDE plot on right tips distribution
```

```python
a, b = plt.subplots(1, 2, figsize=(10, 5))
sns.histplot(tips['total_bill'], kde=True, ax=b[0])
sns.kdeplot(tips['total_bill'], fill=True, ax=b[1])
plt.show()
```

```python
# 2. Show the distribution of size column but as bell shaped curve
sns.kdeplot(tips['size'])
plt.show()
```

```python
# 3. distribution of sepal length grouping by species

iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.histplot(data=iris, x='sepal_length', hue='species')
plt.show()
```

```python
# 4. Make a plot showing distribution of petal length.

sns.histplot(data=iris, x='petal_length')
plt.show()
```
